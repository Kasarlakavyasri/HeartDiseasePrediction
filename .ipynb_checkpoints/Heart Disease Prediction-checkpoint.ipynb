```python
#assigning 1, 2, 3 and 4 to asymptomatic, nonanginal, nontypical and other respectively

for i in range(0,303):
    if(data['ChestPain'][i]=='asymptomatic'):
        data['ChestPain'][i]=1
    elif(data['ChestPain'][i]=='nonanginal'):
         data['ChestPain'][i]=2
    elif(data['ChestPain'][i]=='nontypical'):
        data['ChestPain'][i]=3
    else:
        data['ChestPain'][i]=4
```

```python
#assigning 1 to yes and 0 to no to convert them to numerical values

for i in range(0,303):
    if(data['AHD'][i]=='Yes'):
        data['AHD'][i]=1
    else:
        data['AHD'][i]=0
```

```python
#plotting categorical features

categorical = [('Sex', ['female', 'male']),
               ('ChestPain', ['typical angina', 'atypical angina', 'non-anginal pain', 'asymptomatic']),
               ('Fbs', ['fbs > 120mg', 'fbs < 120mg']),
               ('RestECG', ['normal', 'ST-T wave', 'left ventricular']),
               ('ExAng', ['yes', 'no']),
               ('Slope', ['upsloping', 'flat', 'downsloping'])]

fig_categorial, axes = plt.subplots(nrows=2, ncols=3, figsize=(15, 10))

for i, (feature, categories) in enumerate(categorical):
    row = i // 3
    col = i % 3

    # Generate a color palette for each plot
    colors = sns.color_palette('Set3', len(categories))

    # Plot the bar plot for the categorical feature with different colors for each bar
    ax = axes[row, col]
    ax.bar(data[feature].value_counts().index, data[feature].value_counts().values, color=colors)
    ax.set_title(feature)
    ax.set_xlabel("Categories")
    ax.set_ylabel("Counts")

plt.tight_layout()
plt.show()
```

```python
x=data.iloc[:,:-1]
y=data.iloc[:,-1]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
MinMax=MinMaxScaler()
x_train=MinMax.fit_transform(x_train)
x_test=MinMax.transform(x_test)
y_train=y_train.astype('int')
y_test=y_test.astype('int')
```

```python
#RANDOM FOREST CLASSIFIER

ranfor=RandomForestClassifier(n_estimators=135)
ranfor.fit(x_train,y_train)
y_pred_ranfor=ranfor.predict(x_test)
print('Random Forest Classifier Accuracy Details \n')
print("Accuracy Score :",accuracy_score(y_pred_ranfor,y_test))
print("   f1 Score    :",f1_score(y_pred_ranfor,y_test))
print("Precision Score:",precision_score(y_pred_ranfor,y_test))
print(" Recall Score  :",recall_score(y_pred_ranfor,y_test))
print('Classification Report:\n',classification_report(y_pred_ranfor,y_test))
```

```python
#DECISION TREE CLASSIFIER

dec_tree=DecisionTreeClassifier()
dec_tree.fit(x_train,y_train)
y_pred_dectree=dec_tree.predict(x_test)
print("Decision Tree Accuracy Details:\n")
print("Accuracy Score :",accuracy_score(y_pred_dectree,y_test))
print("   f1 Score    :",f1_score(y_pred_dectree,y_test))
print("Precision Score:",precision_score(y_pred_dectree,y_test))
print(" Recall Score  :",recall_score(y_pred_dectree,y_test))
print('Classification Report:\n',classification_report(y_pred_dectree,y_test))
```

```python
#LOGISTIC REGRESSION

log_reg=LogisticRegression(solver='lbfgs')
log_reg.fit(x_train,y_train)
y_pred_logreg=log_reg.predict(x_test)
print("Logistic Regression Accuracy Details:\n")
print("Accuracy Score :",accuracy_score(y_pred_logreg,y_test))
print("   f1 Score    :",f1_score(y_pred_logreg,y_test))
print("Precision Score:",precision_score(y_pred_logreg,y_test))
print(" Recall Score  :",recall_score(y_pred_logreg,y_test))
print('Classification Report:\n',classification_report(y_pred_logreg,y_test))
```

```python
#KNEIGHBOURS CLASSIFIER

np.random.seed(42)
kneigh=  KNeighborsClassifier()
kneigh.fit(x_train,y_train)
y_pred_kneigh=kneigh.predict(x_test)
print('KNN Classifier Accuracy Details: \n')
print("Accuracy Score :",accuracy_score(y_pred_kneigh,y_test))
print("   f1 Score    :",f1_score(y_pred_kneigh,y_test))
print("Precision Score:",precision_score(y_pred_kneigh,y_test))
print(" Recall Score  :",recall_score(y_pred_kneigh,y_test))
print('Classification Report:\n',classification_report(y_pred_kneigh,y_test))
```

```python
#Naive Bayes classifier

nbay = GaussianNB()
nbay.fit(x_train, y_train)
y_pred_nbay = nbay.predict(x_test)
print("Naive Bayes Accuracy Details:\n")
print("Accuracy Score: ", accuracy_score(y_pred_nbay,y_test))
print("F1 Score: ", f1_score(y_pred_nbay,y_test))
print("Precision Score: ", precision_score(y_pred_nbay,y_test))
print("Recall Score: ", recall_score(y_pred_nbay,y_test))
print("Classification Report:\n", classification_report(y_pred_nbay,y_test))
```

```python
#AdaBoost Classifier

ad_bst = AdaBoostClassifier()
ad_bst.fit(x_train, y_train)
y_pred_adbst = ad_bst.predict(x_test)
print("AdaBoost Accuracy Details:\n")
print("Accuracy Score   :", accuracy_score(y_pred_adbst, y_test))
print("F1 Score         :", f1_score(y_pred_adbst, y_test))
print("Precision Score  :", precision_score(y_pred_adbst, y_test))
print("Recall Score     :", recall_score(y_pred_adbst, y_test))
print("Classification Report:\n", classification_report(y_pred_adbst, y_test))
```

```python
#MLP (neural network classifier)
mlp = MLPClassifier()
mlp.fit(x_train, y_train)
y_pred_mlp = mlp.predict(x_test)
print("Neural Network Accuracy Details:\n")
print("Accuracy Score   :", accuracy_score(y_pred_mlp, y_test))
print("F1 Score         :", f1_score(y_pred_mlp, y_test))
print("Precision Score  :", precision_score(y_pred_mlp, y_test))
print("Recall Score     :", recall_score(y_pred_mlp, y_test))
print("Classification Report:\n", classification_report(y_pred_mlp, y_test))
```
